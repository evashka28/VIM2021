```python
import numpy as np
```

```python
import pandas as pd
DATA = pd.read_csv("student-mat.csv", delimiter=',')
DATA
```

```python
listOfEnum = []

def norm_apply(s) -> int:
  i = listOfEnum.index(s)
  return i

def norm(array):
  return array.unique().tolist()

for column_name in ["school", "sex", "address", "famsize", "Pstatus", "Medu", "Fedu", "Mjob", "Fjob", "reason", "guardian", "schoolsup", "famsup", "paid", "activities", "nursery", "higher", "internet", "romantic"]:
  listOfEnum = norm(DATA[column_name])
  DATA[column_name] = DATA[column_name].apply(norm_apply)  #преобразование столбцов
DATA
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(DATA.drop(["G1","G2","G3"], axis=1).to_numpy(), DATA.G1.to_numpy(), train_size = 0.8)
```

```python
reg = LinearRegression().fit(X_train, y_train) #Обучение модели
reg.score(X_train, y_train) # коэффициент детерминации
```

```python
mean_squared_error(y_test, reg.predict(X_test), squared=False) # Среднеквадратичное отклонение
```

```python
Y_pred = np.apply_along_axis(np.rint,0,reg.predict(X_test))  
Y_pred = list(map(int,Y_pred))    #Предсказание значения столбца
```

```python
from sklearn import metrics                                                             #Характеристики модели
Recall = metrics.recall_score(y_test, Y_pred, average='weighted')                             #Полнота
Precision = metrics.precision_score(y_test, Y_pred, average='weighted', zero_division=0)      #Точность
print("Recall: ", Recall)
print("Precision: ", Precision)
```
